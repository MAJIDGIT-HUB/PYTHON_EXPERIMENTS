```python
#1. Find Maximum and Minimum from a sequence (no built-ins)

def find_max_min(numbers):
    """Returns (max, min) from a sequence without using built-in max/min."""
    if not numbers:
        return None, None

    maximum = numbers[0]
    minimum = numbers[0]

    for num in numbers[1:]:
        if num > maximum:
            maximum = num
        if num < minimum:
            minimum = num

    return maximum, minimum

nums = [10, 6, 8, 90, 12, 56]
max_val, min_val = find_max_min(nums)
print(f"1. Max: {max_val}, Min: {min_val}")
```

```python
# 2. Sum of cubes of all positive integers smaller than n

def sum_of_cubes(n):
    """Returns the sum of cubes of all positive integers < n."""
    total = 0
    for i in range(1, n):      # 1, 2, ..., n-1
        total += i ** 3
    return total

# Test
print(f"2. Sum of cubes below 5: {sum_of_cubes(5)}")
```

```python
# 3. Print 1 to n using recursion (no loops)

def print_1_to_n(current, n):
    """Prints numbers from current up to n recursively."""
    if current > n:
        return
    print(current, end=" ")
    print_1_to_n(current + 1, n)

# Test
print("3. 1 to 7:", end=" ")
print_1_to_n(1, 7)
print()   # newline

```

```python
# 4. Fibonacci series up to n terms (recursive)

def fibonacci(n):
    """Returns the nth Fibonacci number (0-indexed)."""
    if n <= 0:
        return 0
    if n == 1:
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)

def print_fibonacci(n):
    """Prints the Fibonacci series up to n terms."""
    print(f"4. Fibonacci series ({n} terms):", end=" ")
    for i in range(n):
        print(fibonacci(i), end=" ")
    print()

# Test
print_fibonacci(8)
```

```python
# 5. Lambda function to find the volume of a cone ::Formula: V = (1/3) * π * r² * h

import math

volume_of_cone = lambda r, h: (1 / 3) * math.pi * r ** 2 * h

# Test
r, h = 5, 10
print(f"5. Volume of cone (r={r}, h={h}): {volume_of_cone(r, h):.2f}")
```

```python
# 6. Lambda function returning (max, min) tuple from a list

max_min_tuple = lambda lst: (max(lst), min(lst))

# Test
sample = [10, 6, 8, 90, 12, 56]
print(f"6. Max & Min tuple: {max_min_tuple(sample)}")   # (90, 6)
```

```python
# 7. Keyword, Default, and Variable-length Arguments


# a. Keyword Arguments – passed by name, order doesn't matter
def student_info(name, age, grade):
    print(f"  Name: {name}, Age: {age}, Grade: {grade}")

print("7a. Keyword Arguments:")
student_info(age=20, grade="A", name="Alice")   # order changed


# b. Default Arguments – parameter has a preset value if not provided
def greet(name, message="Hello"):
    print(f"  {message}, {name}!")

print("7b. Default Arguments:")
greet("Bob")                  # uses default message
greet("Carol", "Good morning")  # overrides default


# c. Variable-length Arguments
#    *args  → variable positional arguments (tuple)
#    **kwargs → variable keyword arguments (dict)
def variable_args(*args, **kwargs):
    print(f"  Positional args : {args}")
    print(f"  Keyword args    : {kwargs}")

print("7c. Variable-length Arguments:")
variable_args(1, 2, 3, city="Delhi", country="India")


```

```python
# 8. Check whether all values in a dictionary are the same
#    using a lambda function

all_same = lambda d: len(set(d.values())) == 1

# Tests
d1 = {"a": 10, "b": 10, "c": 10}
d2 = {"a": 10, "b": 20, "c": 10}

print(f"8. All values same in {d1}? {all_same(d1)}")   # True
print(f"   All values same in {d2}? {all_same(d2)}")   # False
```

```python
# 9. Create a dictionary from two lists
#    keys → list1,  values → list2

def lists_to_dict(list1, list2):
    """Zips two lists into a dictionary."""
    if len(list1) != len(list2):
        raise ValueError("Both lists must have the same length.")
    return dict(zip(list1, list2))

# Test
keys   = ["name", "age", "city", "country"]
values = ["Diana", 25, "Mumbai", "India"]

result_dict = lists_to_dict(keys, values)
print(f"9. Dictionary from two lists: {result_dict}")
```
